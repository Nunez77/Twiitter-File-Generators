```python
# Define Search Criteria

search_words = ["GME", "AMC"]
date_since = "2021-03-21" # yyyy-mm-dd
date_until = "2021-04-21" # yyyy-mm-dd
language = "en" # Standard language code https://en.wikipedia.org/wiki/List_of_ISO_639-1_codes
country = "USA"  # https://www.iban.com/country-codes
results = "mixed" # for likes and retweets use "popular" and for recent use "recent" use "mixed" for both
how_many_tweets = 100  # 500,000 per month per project
```

```python
# Set up API connection

auth = tweepy.OAuthHandler(consumer_key,
                           consumer_secret)

auth.access_token = access_token
auth.access_token_secret = access_secret

api = tweepy.API(auth,
                 wait_on_rate_limit=True)

places = api.geo_search(query=country,granularity="country")
place_id = places[0].id

# API.search(q[, geocode][, lang][, locale][, result_type][, count][, until][, since_id][, max_id][, include_entities])
tweets = api.search(search_words, count=how_many_tweets, lang=language,
                    result_type=results, since_id=date_since, place=place_id)
```
